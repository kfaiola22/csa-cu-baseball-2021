### 1. Importing and Cleaning Data

```python
import pandas as pd
```

**Pitching Data**

```python
# load cornellpitching20XX.csv from data folder.
pitching2020 = pandas.read_csv('cornellpitching2020.csv')
pitchers2020 = pitching2020[pitching2020['player_id'].notna()]
# .loc[rows,columns] (':' = all), select player rows with at least 1 pitching appearance
appearances2020 = pitchers2020.loc[pitchers2020['App']>0, :]
# col0 redundant, select all rows, cols[1:], index by 'Player'
pitching2020cleaned = appearances2020.iloc[:,1:].set_index('player_id')
# store df in 'pitching20XXcleaned'
pitching2020cleaned
```

```python
# repeat for seasons 2015-2019

# 2019
pitching2019 = pandas.read_csv('cornellpitching2019.csv')
pitchers2019 = pitching2019[pitching2019['player_id'].notna()]
appearances2019 = pitchers2019.loc[pitching2019['App']>0, :]
pitching2019cleaned = appearances2019.iloc[:,1:].set_index('player_id')

#2018
pitching2018 = pandas.read_csv('cornellpitching2018.csv')
pitchers2018 = pitching2018[pitching2018['player_id'].notna()]
appearances2018 = pitchers2018.loc[pitching2018['App']>0, :]
pitching2018cleaned = appearances2018.iloc[:,1:].set_index('player_id')

#2017
pitching2017 = pandas.read_csv('cornellpitching2017.csv')
pitchers2017 = pitching2017[pitching2017['player_id'].notna()]
appearances2017 = pitchers2017.loc[pitching2017['App']>0, :]
pitching2017cleaned = appearances2017.iloc[:,1:].set_index('player_id')

#2016
pitching2016 = pandas.read_csv('cornellpitching2016.csv')
pitchers2016 = pitching2016[pitching2016['player_id'].notna()]
appearances2016 = pitchers2016.loc[pitching2016['App']>0, :]
pitching2016cleaned = appearances2016.iloc[:,1:].set_index('player_id')

#2015
pitching2015 = pandas.read_csv('cornellpitching2015.csv')
pitchers2020 = pitching2015[pitching2015['player_id'].notna()]
appearances2015 = pitchers2020.loc[pitching2015['App']>0, :]
pitching2015cleaned = appearances2015.iloc[:,1:].set_index('player_id')
```

```python
#create totalpitching dataframe containing per player per season data for 2015-2020.
pitchingseasons = [pitching2020cleaned,pitching2019cleaned,pitching2018cleaned,pitching2017cleaned,pitching2016cleaned,pitching2015cleaned]
pitchingsince2015 = pd.concat(pitchingseasons)
pitchingsince2015.to_csv('pitchingsince2015.csv') 
pitchingsince2015
```

**Batting Data**

```python
# load cornellbatting20XX.csv from data folder.
batting2020 = pandas.read_csv('cornellbatting2020.csv')
batters2020 = batting2020[batting2020['player_id'].notna()]
# .loc[rows,columns] (':' = all), select player rows with at least 1 At Bat 
atbats2020 = batters2020.loc[batting2020['BA']>0, :]
# col0 redundant, select all rows, cols[1:], index by 'Player'
batting2020cleaned = atbats2020.iloc[:,1:].set_index('player_id')
# store df in 'batting20XXcleaned'
batting2020cleaned
```

```python
# repeat for seasons 2015-2019

#2019
batting2019 = pandas.read_csv('cornellbatting2019.csv')
batters2019 = batting2019[batting2019['player_id'].notna()]
atbats2019 = batters2019.loc[batters2019['BA']>0, :]
batting2019cleaned = atbats2019.iloc[:,1:].set_index('player_id')

#2018
batting2018 = pandas.read_csv('cornellbatting2018.csv')
batters2018 = batting2018[batting2018['player_id'].notna()]
atbats2018 = batters2018.loc[batters2018['BA']>0, :]
batting2018cleaned = atbats2018.iloc[:,1:].set_index('player_id')

#2017
batting2017 = pandas.read_csv('cornellbatting2017.csv')
batters2017 = batting2017[batting2017['player_id'].notna()]
atbats2017 = batters2017.loc[batters2017['BA']>0, :]
batting2017cleaned = atbats2017.iloc[:,1:].set_index('player_id')

#2016
batting2016 = pandas.read_csv('cornellbatting2016.csv')
batters2016 = batting2016[batting2016['player_id'].notna()]
atbats2016 = batters2016.loc[batting2016['BA']>0, :]
batting2016cleaned = atbats2016.iloc[:,1:].set_index('player_id')

#2015
batting2015 = pandas.read_csv('cornellbatting2015.csv')
batters2015 = batting2015[batting2015['player_id'].notna()]
atbats2015 = batters2015.loc[batting2015['BA']>0, :]
batting2015cleaned = atbats2015.iloc[:,1:].set_index('player_id')
```

```python
#create totalbatting dataframe containing per player per season data for 2015-2020.
battingseasons = [batting2020cleaned,batting2019cleaned,batting2018cleaned,batting2017cleaned,batting2016cleaned,batting2015cleaned]
battingsince2015 = pd.concat(battingseasons)
battingsince2015.to_csv('battingsince2015.csv')
battingsince2015
```
